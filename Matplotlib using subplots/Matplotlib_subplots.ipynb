# Using functions

### We noticed in the last examples that I repeat the same code many times <br> So to avoid so I can make function that will do so for me

```python
def plot_time_series(axes, x, y, my_color, my_xlabel, my_ylabel):
    axes.plot(x, y , color=my_color)
    axes.set_xlabel(my_xlabel,color=my_color)
    axes.set_ylabel(my_ylabel,color=my_color)
    axes.tick_params('y',colors=my_color)
    
## This function will make me plot same figure faster
```

```python
fig, ax = plt.subplots()

plot_time_series(ax, climate_change.index, climate_change['co2'], 'blue', 'Time (years)', 'CO2 levels')
ax2 = ax.twinx()
plot_time_series(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time (years)', 'Relative temperature (Celsius)')

plt.show()
```

# Annotating time series data

### It is usually a small pieces of text that describes a particular part of data

### For the previous graph I will add text to represent data > 1 degree relative temprature

```python
fig, ax = plt.subplots()

plot_time_series(ax, climate_change.index, climate_change['co2'], 'blue', 'Time (years)', 'CO2 levels')
ax2 = ax.twinx()
plot_time_series(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time (years)', 'Relative temperature (Celsius)')

ax2.annotate("more than 1 degree",             ## Text to display
             xy=(("2015-10-06"), 1))           ## coordinates of point I want to display on here (x = 2015-10-06, y = 1)

plt.grid()
plt.show()
# look at point x = 2015-10-06, y = 1 , you will see a written text more than 1
```

#### But the shape written word place is not good so I will change its place

```python
fig, ax = plt.subplots()

plot_time_series(ax, climate_change.index, climate_change['co2'], 'blue', 'Time (years)', 'CO2 levels')
ax2 = ax.twinx()
plot_time_series(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time (years)', 'Relative temperature (Celsius)')

ax2.annotate("more than 1 degree",             ## Text to display
             xy=(("2015-10-06"), 1),          ## coordinates of point I want to display on here (x = 2015-10-06, y = 1)
             xytext=("2010-10-06",0),       ## coordinates of point I want to show text at
             arrowprops = {'arrowstyle':'->','color':'green'}  ## this shows arrow 
            ## note if you made arrowprops = {} it will draw normal arrow without styles
            )  

plt.grid()
plt.show()
# look at point x = 2010-10-06, y = 0 , you will see a written text more than 1
```

## Ok I know that such concept is hard so I will illustrate it again but on sine wave 

### I will create a function named sine_wave that when called It will plot a sine wave function of 1 period from:<br> 0 to 2 pi

```python
pi = np.pi
def sine_wave():
    x = np.linspace(0, 2*pi, 101)
    y = np.sin(x)

    plt.plot(x,y,color='#0398fc')

    plt.title('Sine Wave', color='#0398fc',size=40)
    plt.xlabel('X (rad)',color='#0398fc',size=30)
    plt.ylabel('y',color='#0398fc',size=30)

    plt.xlim([0, 2*pi])             # makes my x-axis from 0 to 2 pi only
    plt.ylim([-1.2, 1.2])

    plt.grid()
```

```python
sine_wave()
plt.show()
```

### Now I want to specify local max and min points which are from math:<br> max: 0.5 pi <br> min: 1.5 pi 

```python
## putting markers at those points
def sine_wave_with_markers():            # function for sine wave with the markers
    sine_wave()
    plt.plot(0.5*pi,1,color='red',marker='o',markersize=5)
    plt.plot(1.5*pi,-1,color='red',marker='o',markersize=5)

sine_wave_with_markers()
plt.show()

## NOTE THAT 
## I could have done that with one function but in a purpose of showing how everything works I made thim in several functions
```

### Now I want to add text and arrows to this points

```python
sine_wave_with_markers()

plt.annotate("Local max point",                               ## Text to display
             xy = (0.5*pi,1),                                 ## point coordinates
             xytext = (pi,1.1),                               ## text coordinates
             arrowprops = {'arrowstyle':'->','color':'red'}   ## arrow
)

plt.annotate("Local min point",                               ## Text to display
             xy = (1.5*pi,-1),                                ## point coordinates
             xytext = (pi,-1.1),                              ## text coordinates
             arrowprops = {'arrowstyle':'->','color':'red'}   ## arrow
)

plt.show()
```

# Histogram

### First I will do some cleaning .....<br>Where I will make new dataframes which are :
> men_rowing (contains data for men in rowing sport)<br>
> men_gymnastics (contains data for men in gymnastics sport)

```python
olympics = pd.read_csv('Datasets/summer2016.csv',index_col='Unnamed: 0')
olympics
```

```python
men_rowing = olympics[(olympics['Sport'] == 'Rowing') & (olympics['Sex'] == 'M')]
men_rowing
```

```python
men_gymnastics = olympics[(olympics['Sport'] == 'Gymnastics') & (olympics['Sex'] == 'M')]
men_gymnastics
```

```python
fig, ax = plt.subplots()
ax.hist(men_rowing['Height'])
ax.hist(men_gymnastics['Height'])
ax.set_xlabel('Height in cm')
ax.set_ylabel('number of observations')
ax.legend(['Rowing','Gymnastics'])
plt.show()
```


### Changing bins

```python
fig, ax = plt.subplots()
ax.hist(men_rowing['Height'],bins=[150,160,170,180,190,200,210])
ax.hist(men_gymnastics['Height'],bins=[150,160,170,180,190,200,210])

ax.set_xlabel('Height in cm')
ax.set_ylabel('number of observations')
ax.legend(['Rowing','Gymnastics'])
plt.show()

## the plot don't look very good so good so I will change histtype to step
```

```python
fig, ax = plt.subplots()
ax.hist(men_rowing['Height'],bins=[150,160,170,180,190,200,210],histtype="step")
ax.hist(men_gymnastics['Height'],bins=[150,160,170,180,190,200,210],histtype="step")

ax.set_xlabel('Height in cm')
ax.set_ylabel('number of observations')
ax.legend(['Rowing','Gymnastics'])
plt.show()
```

# Statistical Plots

## This kind of plots helps me to see statistical summary for my data like mean and std}

```python
fig, ax = plt.subplots()

ax.bar('Rowing',
       men_rowing['Height'].mean(),
       yerr = men_rowing['Height'].std()
      )

ax.bar('Gymnastics',
       men_gymnastics['Height'].mean(),
       yerr = men_gymnastics['Height'].std()
      )

plt.show()

## The shaded area represents the std
```
