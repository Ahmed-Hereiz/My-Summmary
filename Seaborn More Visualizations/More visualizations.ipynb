# Regression Plots

### As you know in ML order matters while fitting.... Some times it is better to fit by 2nd degree so we will learn 2 important functions which are:<br> 
> - regplot()<br>
> - residplot()
### For regplot() it plots the all points of dataset and regression curve fitting them.... <br>While residplot() will be very helpfull to see if your model is good fit or no

```python
bike_data = pd.read_csv('Datasets/bike_share.csv')
bike_data
```

```python
sns.regplot(data=bike_data,x='temp',y='total_rentals')
plt.show()    ## first degree fit
```

### For previous plot residplot() will help me to know if choosing 1st degree fit is good or no

```python
sns.residplot(data=bike_data,x='temp',y='total_rentals')
print(Fore.BLUE + 'The shape her looks to have slight curve')
plt.show()

# data should be randomly plotted arround mid line but here data looks like it has slight curve 
# so I will see residplot() with order=2 to see if it is better

sns.residplot(data=bike_data,x='temp',y='total_rentals',order=2,color='g')
print(Fore.GREEN + 'The shape her looks more randomly distributed which is better ')
print(Fore.GREEN + 'So it will be better to use 2nd degree in fitting')
plt.show()
```

```python
sns.regplot(data=bike_data,x='temp',y='total_rentals',order=2,color='g')
plt.show()    ## second degree fit
```

### And this looks much better so our degree will be 2nd degree

## Plotting with Categorical values

```python
sns.regplot(data=bike_data,x='mnth',y='total_rentals',order=2)
print(Fore.BLUE + 'Without jitter:')
plt.show()   

## x_jitter can make it more randomized

sns.regplot(data=bike_data,x='mnth',y='total_rentals',order=2, x_jitter=0.1,color='y')
print(Fore.YELLOW + 'With little jitter:')
plt.show()  

sns.regplot(data=bike_data,x='mnth',y='total_rentals',order=2, x_jitter=0.5,color='r')
print(Fore.RED + 'With large jitter:')
plt.show()   
```

## Using Estimator
#### Used for highligiting trends

```python
sns.regplot(data=bike_data,x='mnth',y='total_rentals',order=2,x_estimator=np.mean)
print(Fore.BLUE + 'estimator uses mean:')
plt.show()   

sns.regplot(data=bike_data,x='mnth',y='total_rentals',order=2,x_estimator=np.median,color='r')
print(Fore.RED + 'estimator uses median:')
plt.show()   
```

## We can even divide this data into bins:

```python
sns.regplot(data=bike_data,x='temp',y='total_rentals',order=1,x_bins=4)
plt.show()

sns.regplot(data=bike_data,x='temp',y='total_rentals',order=2,x_bins=4)
plt.show()
```

# Matrix Plot
### To do a heatmap() we will need a matrix of data so we will use crosstab() from pandas to do so

```python
bike_data
```

```python
to_heat_map = pd.crosstab(bike_data['mnth'],        # mnth will be on rows (take average of total_rentals)
            bike_data['weekday'],             # weekday will be on cols (take average of total_rentals)
            values=bike_data['total_rentals'],  # making our values from total_rentals
            aggfunc='mean'                         # taking mean of total_rentals
           ).round(0)                              # making value rounded

to_heat_map
```

### Now this shape is ready for being a heatmap

```python
sns.heatmap(to_heat_map)
plt.show()
```

## Customizing Heatmaps

```python
to_heat_map = to_heat_map.astype('int')    # changing data to int 

sns.heatmap(to_heat_map,           # data to plot
           annot=True,             # makes annotating True
           fmt='d',                # results will be integers
           cmap='ocean',           # cmap (makes colors for cells)
           cbar=True,              # Vertical bar   
           linewidths=0.1,         
           )
plt.show()

## ------------------------------------------------------------------ ## 

## changing in formats:

sns.heatmap(to_heat_map,           # data to plot
           annot=False,             # makes annotating True
           fmt='d',                # results will be integers
           cmap='crest',           # cmap (makes colors for cells)
           cbar=False,              # Vertical bar   
           linewidths=0.6,         
           )
print('There is no numbers because annot=false')
plt.show()

## ------------------------------------------------------------------ ## 

## changing in formats:

sns.heatmap(to_heat_map,           # data to plot
           annot=True,             # makes annotating True
           fmt='d',                # results will be integers
           cmap='Set3',           # cmap (makes colors for cells)
           cbar=False,              # Vertical bar   
           linewidths=10,         
           )
print('Look how line width increased between cells')
plt.show()
```

### Heatmap is also used in representing correlations between columns

```python
my_cols = ['total_rentals', 'temp', 'casual', 'hum', 'windspeed']
sns.heatmap(bike_data[my_cols].corr(),cmap='Blues')
plt.show()
```

# Facetgrid

#### One of Seaborn's most powerful features is its ability to combine multiple smaller plots into a larger visualization that can help identify trends in data with many variables.
#### Seaborn's FacetGrid manages the back end data manipulations to make sure that the data is split across rows, columns, and hue, and then used to make the appropriate plot type. The key point to remember is that FacetGrid() provides a lot of flexibility, but you must use a two step process of defining the Facets and mapping the plot type.

```python
pd.set_option('display.max_columns', None)

college_data = pd.read_csv('Datasets/college_datav3.csv')
college_data
```

## FacetGrid for categorical plot

```python
g = sns.FacetGrid(data=college_data,col='HIGHDEG')
g.map(sns.boxplot,'Tuition',order=['0','1','2','3','4'])

plt.show()

# I don't know why they use word g while using FacetGrid... so I followed the conventions
```

#### Using map function on Facetgrid I made many plots

## Catplots (Was previously illustrated in the previous Notebook)

#### Seaborn's FacetGrid() is very powerful and flexible but involves multiple steps to create. The catplot() function is a shortcut to creating FacetGrids. The underlying returned value is a FacetGrid but the process for creating one is much simpler. The single catplot() function takes care of the two step process for you.

```python
sns.catplot(x='Tuition',col='HIGHDEG',data=college_data,kind='box')

plt.show()
```

## FacetGrid for regression plot

```python
g = sns.FacetGrid(data=college_data,col='HIGHDEG')
g.map(plt.scatter,'Tuition','SAT_AVG_ALL')

plt.show()
```

## lmplots

#### The lmplot() function is similar to the catplot() function. It provides a shortcut for plotting regression and scatter plots on FacetGrids. In this example, we create a plot that is similar to the FacetGrid() scatter plot. We have also disabled regression lines with the fit_reg equals False parameter.

```python
sns.lmplot(data=college_data,x='Tuition',y='SAT_AVG_ALL',col='HIGHDEG',fit_reg=False) # fit_reg = False

plt.show()
# Which is same as before

## --------------------------------------------------------------------- ##
sns.lmplot(data=college_data,x='Tuition',y='SAT_AVG_ALL',col='HIGHDEG',fit_reg=True) # fit_reg = True

plt.show()
```

# PairGrid

####  PairGrid and pairplot. These plots are similar to the FacetGrid, catplot, and lmplots
#### The PairGrid plot allows us to build plots that show the relationships between two data elements

```python
insurance.head()
```

## Making pairgrids

```python
g = sns.PairGrid(insurance, vars=["fatal_collisions", "premiums"])
g2 = g.map(sns.scatterplot)

plt.show()
```

### Here Iam specifing the diagonal

```python
g = sns.PairGrid(insurance, vars=["fatal_collisions", "premiums"])
g2 = g.map_diag(sns.histplot)              # main diagonal
g3 = g2.map_offdiag(sns.scatterplot)

plt.show()
```

## Pairplots

#### Pairplot is a shortcut function that simplifies the PairGrid. First, we define the variables like we did in the previous view. However, instead of mapping the plots, we can pass keywords to the function to specify the kind of plots for the main and off diagonals. The other nice aspect of this function is that it plots a regression line for us. This is a useful feature for visualizing potential relationships in the data.

```python
sns.pairplot(data=insurance,
        vars=["fatal_collisions", "premiums"],
        kind='scatter')

plt.show()
```

```python
sns.pairplot(data=insurance,
        vars=["fatal_collisions", "premiums"],
        kind='scatter',
        hue='Region',
        palette='RdBu',
        diag_kws={'alpha':.5})

plt.show()
```

```python
sns.pairplot(data=insurance,
        x_vars=["fatal_collisions_speeding", "fatal_collisions_alc"],
        y_vars=['premiums', 'insurance_losses'],
        kind='scatter',
        hue='Region',
        palette='husl')

plt.show()

# here Iam using 4 data columns to compare with
```

```python
sns.pairplot(data=insurance,
             vars=["insurance_losses", "premiums"],
             kind='reg',
             palette='BrBG',
             diag_kind = 'kde',
             hue='Region')

plt.show()
```
