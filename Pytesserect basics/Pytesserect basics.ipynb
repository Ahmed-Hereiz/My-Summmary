## There is till a problem with pytesserect... that it is bad with images with noise :''

```python
text_image_original = Image.open('Images/text_image3.png')
text_image = text_image_original.resize((750,250))   
my_text_image = ImageOps.expand(text_image,border=5,fill='black')
display(my_text_image)

text = pytesseract.image_to_string(text_image)
print(text)
```

### We can note here that there was noise in the background so... we couldn't read well.

### As a solution for such thing we will need to change background color to grey :

```python
img = Image.open('Images/text_image3.png')
img = img.convert('L')

## I will save new image as greyscale
img.save('Images/greyscale.jpg')

## Now let's see how will pytesserect see the new image : 
text_image_original = Image.open('Images/greyscale.jpg')
text_image = text_image_original.resize((750,250))   
my_text_image = ImageOps.expand(text_image,border=5,fill='black')
display(my_text_image)

text = pytesseract.image_to_string(text_image)
print(text)
```

## Now it saw it correctly :)

### Ok now let's work on a different example : 

```python
text_image_original = Image.open('Images/text_image4.png')
text_image = text_image_original.resize((750,350))   
my_text_image = ImageOps.expand(text_image,border=5,fill='black')
display(my_text_image)
```

```python
text = pytesseract.image_to_string(text_image)
print(text)
```

## pytesserect can read nothing here :(

### As a solution... I will crop the image.

```python
bounding_box = (190, 50, 490, 140)

title_image=my_text_image.crop(bounding_box)
display(title_image)

text = pytesseract.image_to_string(title_image)
print(text)
```
