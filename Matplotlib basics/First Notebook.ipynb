# piechart
plt.pie()

```python
data_value = [60, 30, 28, 78, 99]
data_labels = ['A', 'B', 'C', 'D', 'F']
data_colors = ['#3082cf', '#6030cf', '#30cfb5', '#30cf50', '#b430df']

plt.pie(data_value, labels=data_labels, colors=data_colors, wedgeprops={'edgecolor':'black'})
# edge color adds black edges check more in documentation
plt.show()
```

## What If you want to see specific data like A 
### Here we will use explode parameter

```python
plt.title('My pie chart :) ')

my_explode = [0.2, 0, 0, 0, 0]    # make all 0 exept data you want to show
plt.pie(data_value, labels=data_labels, colors=data_colors,explode=my_explode , wedgeprops={'edgecolor':'black'})
plt.show()

## ----------------------------------------------------------------- ##

# You can also add shadows

plt.pie(data_value, labels=data_labels, colors=data_colors,explode=my_explode , wedgeprops={'edgecolor':'black'}, 
        shadow=True)
plt.show()

## ----------------------------------------------------------------- ##

# you can change start angles

plt.pie(data_value, labels=data_labels, colors=data_colors,explode=my_explode , wedgeprops={'edgecolor':'black'}, 
        shadow=True, startangle=90)
plt.show()
```

### to add % on the charts:

```python
plt.pie(data_value, labels=data_labels, colors=data_colors,explode=my_explode , wedgeprops={'edgecolor':'black'}, 
        shadow=True, startangle=90, autopct='%1.1f%%')  ## And you can know this from documentation
plt.show()
```

# Visualizing data using subplots


```python
df = pd.read_csv('pokemon.csv',index_col='id')
df
```

### I will choose some rows for a dataframe which I will work on ....<br> That is because I don't nead the whole dataframe

```python
my_df = df[['species', 'height', 'weight', 'base_experience','type_1','hp','attack','defense','speed']]
my_df
```

### This is the old way when plotting many variables
### When using subplots you can make subplot for every figure so we use subplot() method

```python
height = sorted(my_df['height'])
attack = sorted(my_df['attack'])
defense = sorted(my_df['defense'])
speed = sorted(my_df['speed'])
hp = sorted(my_df['hp'])

plt.plot(height,attack,color='#2d98d6')
plt.plot(height,defense,color='#2dd66e')
plt.plot(height,speed,color='#7c2dd6')
plt.plot(height,hp,color='#d62d2d')

plt.title('Pokemon data')
plt.xlabel('height of pokemon')
plt.ylabel('pokemon data')
plt.legend(['weight','defense','speed','hp'])

plt.show()
```

### Now I will do same example using subplot

```python
fig, ax = plt.subplots()

ax.plot(height,attack,color='#2d98d6')
ax.plot(height,defense,color='#2dd66e')
ax.plot(height,speed,color='#7c2dd6')
ax.plot(height,hp,color='#d62d2d')

ax.set_title('Pokemon data')             # not that it changed to ax.set_title
ax.set_xlabel('height of pokemon')       # not that it changed to ax.set_xlabel
ax.set_ylabel('pokemon data')            # not that it changed to ax.set_ylabel
ax.legend(['weight','defense','speed','hp'])

plt.show()
```

### It gave same result as before and that is because plt.subplot() when you don't give it tuple of variables it consider the shape to be 1 subplot

```python
fig, ax = plt.subplots(nrows=3,ncols=2)
# now I made empty subplots
```

```python
print(ax)        
print(ax[0][0])
## or I can uupack data 
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,ncols=1)
# note (ax1, ax2) number must be equal number of rows to unpack
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,ncols=2)

ax1[0].plot(height,attack,color='#2d98d6')   # Iam Indexing avery figure to make the plot in it
ax1[1].plot(height,defense,color='#2dd66e')
ax2[0].plot(height,speed,color='#7c2dd6')
ax2[1].plot(height,hp,color='#d62d2d')

ax1[0].set_title('Pokemon data') 
ax1[0].set_xlabel('height of pokemon')       
ax1[0].set_ylabel('pokemon data')            
ax1[0].legend(['weight','defense','speed','hp'])

for i in range(2):           # 2 is number of columns and for loop for making labels and title
    ax1[i].set_title('Pokemon data') 
    ax1[i].set_xlabel('height of pokemon')       
    ax1[i].set_ylabel('pokemon data')            
    ax1[i].legend(['weight','defense','speed','hp'])
    
    ax2[i].set_title('Pokemon data') 
    ax2[i].set_xlabel('height of pokemon')       
    ax2[i].set_ylabel('pokemon data')            
    ax2[i].legend(['weight','defense','speed','hp'])

plt.show()
```
